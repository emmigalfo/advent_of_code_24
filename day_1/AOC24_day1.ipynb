### Part 1

```python
import pandas as pd
import numpy as np
```

```python
cols = ['first', 'second']
df = pd.read_csv('/Users/emmi/Desktop/aoc_day1_1.csv', header=None, names=cols)
```

```python
df.head()
```

```python
sorted_first = df['first'].sort_values().reset_index(drop=True)
sorted_second = df['second'].sort_values().reset_index(drop=True)
sorted_df = pd.DataFrame({'first': sorted_first, 'second': sorted_second})
```

```python
sorted_df.head()
```

```python
sorted_df['difference'] = abs(sorted_df['first'] - sorted_df['second'])
```

```python
sorted_df.head()
```

```python
result = sum(sorted_df['difference'])
```

```python
result
```

### Part 2

```python
right_counts = sorted_second.value_counts().to_dict()
right_counts
```

```python
similarity_score = 0
for num in sorted_first:
    count_in_right = right_counts.get(num, 0)
    similarity_score += num * count_in_right
similarity_score
```
